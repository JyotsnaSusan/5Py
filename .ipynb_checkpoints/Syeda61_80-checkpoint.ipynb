####64. Plot a graph by counting the no. of patients in each department.

# 66. Plot a graph showing the count of patients for the reason visited the hospital.

```python
df_size = df_EDvisits.groupby(['ReasonForVisit'])['PatientID'].count()
data = df_size.to_dict()
dd = list(data.keys())
count = list(data.values())
#print(df_size)

plt.pie(count, labels= df_size)
#plt.show()

plt.title('Count of patients for the reason visited the hospital')

# Add Legends
plt.legend(dd, bbox_to_anchor=(1.05, 1.0), loc="upper left", fontsize='8')

# show the graph
plt.show()
```

# 67. Create a subplot on x = np.arange(0, 10, 0.1) , y = np.sin(np.pi * x) + x with 2 rows & 2 columns with object oriented approach.

```python
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.1)
y = np.sin(np.pi * x) + x

fig, axs = plt.subplots(nrows=2 , ncols=2, constrained_layout=True)

fig.suptitle ('Sine wave' , fontsize=12)

axs[0][0].plot(x,y,'y-')
axs[0][1].plot(x,y,'g o')
axs[1][0].plot(x,y,'b *')
axs[1][1].plot(x,y,'k ^')


for ax in axs.flat:
    ax.set(xlabel='Angle', ylabel='Sine')

plt.show()
```

# 70.Create a visualization on the count of patients based on primary diagnosis.

```python
df_size = df_ReadmissionRegistry.groupby(['PrimaryDiagnosis'])['PatientID'].count()
data = df_size.to_dict()
dd = list(data.keys())
count = list(data.values())
#print(df_size)

plt.pie(count, labels= df_size)
centre_circle = plt.Circle((0, 0), 0.50, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Visits By Primary Diagnosis')

# Add Legends
plt.legend(dd, bbox_to_anchor=(1.05, 1.0), loc="upper left", fontsize='8')

# show the graph
plt.show()
```
